Data Visualization:

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Create bar chart for 'type'
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=netflix_data, hue='type', palette='pastel', legend=False)
plt.title('Distribution of Movie vs TV Show')
plt.xlabel('Type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Create histogram for 'release_year'
plt.figure(figsize=(12, 6))
plt.hist(netflix_data['release_year'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Plot the distribution of movie durations
plt.figure(figsize=(10, 6))
sns.histplot(data=netflix_data[netflix_data['type'] == 'Movie'], x='duration', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Create histogram for 'duration' for TV shows
plt.figure(figsize=(12, 6))
tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show']
plt.hist(tv_shows_data['duration'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot to explore relationship between 'release_year' and 'rating'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='release_year', y='rating', data=netflix_data)
plt.title('Relationship between Release Year and Rating')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
# Get top 10 countries with most content
top_countries = netflix_data['country'].value_counts().head(10)

# Create bar chart for top countries with most content
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, color='skyblue')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Clustering Analysis
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Convert 'duration' column to float if not already in that format
netflix_data['duration'] = netflix_data['duration'].astype(float)

# Select features for clustering
features = netflix_data[['release_year', 'duration']].dropna()  # Drop rows with missing values

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)

# Assign cluster labels to the original dataset
features['cluster'] = cluster_labels
netflix_data_clustered = pd.concat([netflix_data.drop(columns=['release_year', 'duration']), features], axis=1)

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='release_year', y='duration', hue='cluster', data=netflix_data_clustered, palette='tab10')
plt.title('K-Means Clustering of Netflix Content')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.grid(True)
plt.show()
```
